# NYC Housing Affordability — Visual Analysis
**NTA-Level Panel Dataset (2012–2022) | 2,512 observations | 239 neighborhoods**

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from scipy.stats import gaussian_kde

plt.rcParams["figure.dpi"] = 110
plt.rcParams["font.size"] = 11
sns.set_style("whitegrid")

df = pd.read_csv("expanded_data/nta_panel_final.csv")
print(f"Dataset: {df.shape[0]} rows x {df.shape[1]} columns")
print(f"Boroughs: {sorted(df['borough_name'].unique().tolist())}")
print(f"Years: {sorted(df['year'].unique().tolist())}")
print(f"NTAs: {df['nta_code'].nunique()} unique neighborhoods")
df.head(3)
```

## Fig 1 — Distribution of Rent Burden

```python
col = "rent_burden_30plus_pct"
data_clean = df[col].dropna()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(data_clean, bins=40, color="steelblue", edgecolor="white", alpha=0.85)
axes[0].axvline(data_clean.mean(), color="red", linestyle="--", lw=2, label=f"Mean: {data_clean.mean():.3f}")
axes[0].axvline(data_clean.median(), color="orange", linestyle="--", lw=2, label=f"Median: {data_clean.median():.3f}")
axes[0].set_xlabel("Rent Burden 30%+"); axes[0].set_ylabel("Frequency")
axes[0].set_title("Histogram of Rent Burden (30%+ Threshold)"); axes[0].legend()

kde = gaussian_kde(data_clean)
x = np.linspace(data_clean.min(), data_clean.max(), 300)
axes[1].plot(x, kde(x), color="steelblue", lw=2)
axes[1].fill_between(x, kde(x), alpha=0.3, color="steelblue")
axes[1].axvline(data_clean.mean(), color="red", linestyle="--", lw=2, label=f"Mean: {data_clean.mean():.3f}")
axes[1].axvline(data_clean.median(), color="orange", linestyle="--", lw=2, label=f"Median: {data_clean.median():.3f}")
axes[1].set_xlabel("Rent Burden 30%+"); axes[1].set_ylabel("Density")
axes[1].set_title("KDE of Rent Burden (30%+ Threshold)"); axes[1].legend()

plt.suptitle("Fig 1: Rent Burden Distribution Across All NTAs (2012-2022)", fontsize=13, y=1.02)
plt.tight_layout(); plt.show()
```

## Fig 2 — Severe Rent Burden by Borough

```python
borough_order = sorted(df["borough_name"].dropna().unique())
plot_data = [df.loc[df["borough_name"]==b, "rent_burden_50plus_pct"].dropna().values for b in borough_order]
colors = ["#4472C4", "#ED7D31", "#A9D18E", "#FF4444"]

fig, ax = plt.subplots(figsize=(11, 6))
bp = ax.boxplot(plot_data, labels=borough_order, patch_artist=True,
                medianprops=dict(color="black", linewidth=2.5))
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color); patch.set_alpha(0.75)

ax.set_xlabel("Borough", fontsize=12)
ax.set_ylabel("Severely Cost-Burdened Share (50%+)", fontsize=12)
ax.set_title("Fig 2: Severe Rent Burden Distribution by Borough (2012-2022)", fontsize=13)
ax.yaxis.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout(); plt.show()
```

## Fig 3 — Rent Burden Trends Over Time

```python
trend30 = df.groupby(["borough_name","year"])["rent_burden_30plus_pct"].mean().reset_index()
trend50 = df.groupby(["borough_name","year"])["rent_burden_50plus_pct"].mean().reset_index()
colors_b = ["#4472C4","#ED7D31","#A9D18E","#FF4444"]
boroughs = sorted(df["borough_name"].unique())

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
for i, boro in enumerate(boroughs):
    d30 = trend30[trend30["borough_name"]==boro].sort_values("year")
    d50 = trend50[trend50["borough_name"]==boro].sort_values("year")
    axes[0].plot(d30["year"], d30["rent_burden_30plus_pct"], marker="o", lw=2, markersize=5, color=colors_b[i], label=boro)
    axes[1].plot(d50["year"], d50["rent_burden_50plus_pct"], marker="o", lw=2, markersize=5, color=colors_b[i], label=boro)

for ax, title, ylab in zip(axes,
    ["Rent Burden 30%+ Trend", "Severe Rent Burden 50%+ Trend"],
    ["Mean Rent Burden (30%+)", "Mean Severe Rent Burden (50%+)"]):
    ax.axvspan(2019.5, 2021.5, alpha=0.15, color="red", label="COVID-19")
    ax.axvline(2020, color="red", linestyle=":", alpha=0.5, lw=1.5)
    ax.set_xlabel("Year"); ax.set_ylabel(ylab)
    ax.set_title(title, fontsize=12)
    ax.legend(fontsize=9)
    ax.set_xticks(sorted(df["year"].unique()))
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.suptitle("Fig 3: Rent Burden Trends by Borough (2012-2022)", fontsize=13, y=1.02)
plt.tight_layout(); plt.show()
```

## Fig 4 — Top 20 Most Burdened Neighborhoods (Heatmap)

```python
nta_avg = df.groupby("nta_name")["rent_burden_50plus_pct"].mean().dropna()
top20 = nta_avg.nlargest(20).index.tolist()

pivot = df[df["nta_name"].isin(top20)].pivot_table(
    index="nta_name", columns="year", values="rent_burden_50plus_pct", aggfunc="mean")
pivot = pivot.loc[nta_avg[top20].sort_values(ascending=False).index]

fig, ax = plt.subplots(figsize=(14, 8))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlOrRd",
            linewidths=0.4, linecolor="white",
            cbar_kws={"label": "Severely Cost-Burdened Share"}, ax=ax)
ax.set_title("Fig 4: NTA-Level Severe Rent Burden Heatmap — Top 20 Most Burdened Neighborhoods", fontsize=13)
ax.set_xlabel("Year"); ax.set_ylabel("Neighborhood (NTA)")
plt.xticks(rotation=45); plt.yticks(rotation=0)
plt.tight_layout(); plt.show()
```

## Fig 5 — Worst Neighborhoods in 2022 by Borough

```python
year_2022 = df[df["year"]==2022].dropna(subset=["rent_burden_50plus_pct"])
colors_boro = {"Bronx":"#4472C4","Brooklyn":"#ED7D31","Manhattan":"#A9D18E","Queens":"#FF4444"}

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, boro in enumerate(sorted(year_2022["borough_name"].unique())):
    top10 = (year_2022[year_2022["borough_name"]==boro]
             .sort_values("rent_burden_50plus_pct", ascending=False).head(10))
    color = colors_boro.get(boro, "gray")
    axes[i].barh(top10["nta_name"][::-1], top10["rent_burden_50plus_pct"][::-1],
                 color=color, edgecolor="white", alpha=0.85)
    axes[i].set_xlabel("Severely Cost-Burdened Share (50%+)", fontsize=10)
    axes[i].set_title(f"{boro} — Top 10 Most Burdened NTAs (2022)", fontsize=11, color=color)
    axes[i].set_xlim(0, year_2022["rent_burden_50plus_pct"].max() * 1.1)
    for j, (_, row) in enumerate(top10[::-1].iterrows()):
        axes[i].text(row["rent_burden_50plus_pct"] + 0.003, j,
                     f"{row['rent_burden_50plus_pct']:.3f}", va="center", fontsize=8)

plt.suptitle("Fig 5: Top 10 Most Severely Rent-Burdened NTAs per Borough (2022)", fontsize=14, y=1.01)
plt.tight_layout(); plt.show()
```

## Fig 6 — Income vs Rent Burden Scatter

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for i, boro in enumerate(sorted(df["borough_name"].unique())):
    sub = df[df["borough_name"]==boro]
    color = colors_boro.get(boro, "gray")
    axes[0].scatter(sub["median_hh_income"]/1000, sub["rent_burden_50plus_pct"],
                    alpha=0.25, s=12, color=color, label=boro)
    axes[1].scatter(sub["median_gross_rent"], sub["rent_burden_50plus_pct"],
                    alpha=0.25, s=12, color=color, label=boro)

axes[0].set_xlabel("Median Household Income (/bin/zshs)")
axes[0].set_ylabel("Severely Cost-Burdened Share (50%+)")
axes[0].set_title("Income vs. Severe Rent Burden")
axes[0].legend(fontsize=9)

axes[1].set_xlabel("Median Gross Rent ($)")
axes[1].set_ylabel("Severely Cost-Burdened Share (50%+)")
axes[1].set_title("Rent Level vs. Severe Rent Burden")
axes[1].legend(fontsize=9)

plt.suptitle("Fig 6: Key Drivers of Severe Rent Burden (All NTAs, 2012-2022)", fontsize=13, y=1.02)
plt.tight_layout(); plt.show()
```

## Fig 7 — Correlation Heatmap

```python
corr_cols = ["median_hh_income","median_gross_rent","rent_burden_30plus_pct",
             "rent_burden_50plus_pct","vacancy_rate","unemployment_rate",
             "severe_crowding_rate","gini_coefficient","renter_share",
             "homeownership_rate","eviction_rate"]
available = [c for c in corr_cols if c in df.columns]
corr = df[available].corr().round(2)

fig, ax = plt.subplots(figsize=(13, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="RdBu_r",
            center=0, vmin=-1, vmax=1, linewidths=0.4,
            cbar_kws={"label": "Pearson Correlation"}, ax=ax)
ax.set_title("Fig 7: Correlation Matrix of Key Housing Variables", fontsize=13)
plt.xticks(rotation=45, ha="right"); plt.yticks(rotation=0)
plt.tight_layout(); plt.show()
```

## Fig 8 — Neighborhoods Most Improved and Most Worsened (2012–2022)

```python
nta_2012 = df[df["year"]==2012].set_index("nta_name")["rent_burden_50plus_pct"]
nta_2022 = df[df["year"]==2022].set_index("nta_name")["rent_burden_50plus_pct"]
# Convert to dict to avoid Series-as-key errors
boro_map_dict = (df[df["year"]==2022]
                 .drop_duplicates("nta_name")
                 .set_index("nta_name")["borough_name"]
                 .to_dict())

common = nta_2012.index.intersection(nta_2022.index)
change = (nta_2022[common] - nta_2012[common]).dropna().sort_values()

top10_improved = change.head(10)
top10_worsened = change.tail(10)

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

colors_imp = [colors_boro.get(boro_map_dict.get(n, ""), "steelblue") for n in top10_improved.index]
axes[0].barh(range(len(top10_improved)), top10_improved.values, color=colors_imp, alpha=0.85, edgecolor="white")
axes[0].set_yticks(range(len(top10_improved)))
axes[0].set_yticklabels(top10_improved.index, fontsize=9)
axes[0].axvline(0, color="black", lw=1.2)
axes[0].set_xlabel("Change in Severe Rent Burden (2012 to 2022)")
axes[0].set_title("Top 10 Most Improved NTAs", fontsize=12, color="green")

colors_wor = [colors_boro.get(boro_map_dict.get(n, ""), "steelblue") for n in top10_worsened.index]
axes[1].barh(range(len(top10_worsened)), top10_worsened.values, color=colors_wor, alpha=0.85, edgecolor="white")
axes[1].set_yticks(range(len(top10_worsened)))
axes[1].set_yticklabels(top10_worsened.index, fontsize=9)
axes[1].axvline(0, color="black", lw=1.2)
axes[1].set_xlabel("Change in Severe Rent Burden (2012 to 2022)")
axes[1].set_title("Top 10 Most Worsened NTAs", fontsize=12, color="red")

patches = [mpatches.Patch(color=v, label=k) for k,v in colors_boro.items()]
fig.legend(handles=patches, loc="lower center", ncol=4, fontsize=10, bbox_to_anchor=(0.5, -0.04))
plt.suptitle("Fig 8: Neighborhood Change in Severe Rent Burden 2012 to 2022", fontsize=14, y=1.02)
plt.tight_layout(); plt.show()
```

## Fig 9 — Borough Rent & Income Divergence

```python
boro_year = df.groupby(["borough_name","year"]).agg(
    rent=pd.NamedAgg("median_gross_rent","mean"),
    income=pd.NamedAgg("renter_median_income","mean")).reset_index()

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, boro in enumerate(sorted(boro_year["borough_name"].unique())):
    sub = boro_year[boro_year["borough_name"]==boro].sort_values("year")
    color = colors_boro.get(boro, "gray")
    ax = axes[i]
    ax2 = ax.twinx()
    l1 = ax.plot(sub["year"], sub["rent"], color=color, marker="o", lw=2.2, markersize=5, label="Median Gross Rent")
    l2 = ax2.plot(sub["year"], sub["income"]/1000, color="navy", marker="s", lw=2, linestyle="--", markersize=5, label="Renter Income (/bin/zshs)")
    ax.axvspan(2019.5, 2021.5, alpha=0.1, color="red")
    ax.set_xlabel("Year"); ax.set_ylabel("Median Gross Rent ($)", color=color)
    ax2.set_ylabel("Renter Median Income (/bin/zshs)", color="navy")
    ax.set_title(f"{boro}", fontsize=12)
    ax.set_xticks(sub["year"]); plt.setp(ax.get_xticklabels(), rotation=45, fontsize=8)
    lines = l1 + l2
    ax.legend(lines, [l.get_label() for l in lines], fontsize=8, loc="upper left")

plt.suptitle("Fig 9: Rent vs. Renter Income Divergence by Borough (2012-2022)",
             fontsize=13, y=1.01)
plt.tight_layout(); plt.show()
```

## Fig 10 — Eviction Rate vs Rent Burden

```python
if "eviction_rate" in df.columns:
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    evic_trend = df.groupby(["borough_name","year"])["eviction_rate"].mean().reset_index()
    for i, boro in enumerate(sorted(df["borough_name"].unique())):
        sub = evic_trend[evic_trend["borough_name"]==boro].sort_values("year")
        axes[0].plot(sub["year"], sub["eviction_rate"], marker="o", lw=2,
                     color=colors_b[i], label=boro, markersize=5)
    axes[0].axvspan(2019.5, 2021.5, alpha=0.15, color="red", label="COVID-19")
    axes[0].set_xlabel("Year"); axes[0].set_ylabel("Eviction Rate (per 1,000 renters)")
    axes[0].set_title("Eviction Rate Trend by Borough"); axes[0].legend(fontsize=9)
    axes[0].set_xticks(sorted(df["year"].unique())); plt.setp(axes[0].get_xticklabels(), rotation=45)

    sub2 = df[["eviction_rate","rent_burden_50plus_pct","borough_name"]].dropna()
    for i, boro in enumerate(sorted(sub2["borough_name"].unique())):
        s = sub2[sub2["borough_name"]==boro]
        axes[1].scatter(s["eviction_rate"], s["rent_burden_50plus_pct"],
                        alpha=0.3, s=15, color=colors_b[i], label=boro)
    axes[1].set_xlabel("Eviction Rate (per 1,000)"); axes[1].set_ylabel("Severely Cost-Burdened Share")
    axes[1].set_title("Eviction Rate vs. Severe Rent Burden"); axes[1].legend(fontsize=9)

    plt.suptitle("Fig 10: Eviction Analysis (NYC NTAs 2012-2022)", fontsize=13, y=1.02)
    plt.tight_layout(); plt.show()
else:
    print("eviction_rate column not found in dataset")
```

## Fig 11 — Income Inequality (Gini Coefficient)

```python
if "gini_coefficient" in df.columns:
    gini_trend = df.groupby(["borough_name","year"])["gini_coefficient"].mean().reset_index()
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    for i, boro in enumerate(sorted(gini_trend["borough_name"].unique())):
        sub = gini_trend[gini_trend["borough_name"]==boro].sort_values("year")
        axes[0].plot(sub["year"], sub["gini_coefficient"], marker="o", lw=2,
                     color=colors_b[i], label=boro, markersize=5)
    axes[0].axvspan(2019.5, 2021.5, alpha=0.15, color="red", label="COVID-19")
    axes[0].set_xlabel("Year"); axes[0].set_ylabel("Mean Gini Coefficient")
    axes[0].set_title("Income Inequality Trend by Borough"); axes[0].legend(fontsize=9)
    axes[0].set_xticks(sorted(df["year"].unique())); plt.setp(axes[0].get_xticklabels(), rotation=45)

    sub2 = df[["gini_coefficient","rent_burden_50plus_pct","borough_name"]].dropna()
    for i, boro in enumerate(sorted(sub2["borough_name"].unique())):
        s = sub2[sub2["borough_name"]==boro]
        axes[1].scatter(s["gini_coefficient"], s["rent_burden_50plus_pct"],
                        alpha=0.3, s=15, color=colors_b[i], label=boro)
    axes[1].set_xlabel("Gini Coefficient"); axes[1].set_ylabel("Severely Cost-Burdened Share")
    axes[1].set_title("Inequality vs. Severe Rent Burden"); axes[1].legend(fontsize=9)

    plt.suptitle("Fig 11: Income Inequality & Housing Affordability (NYC 2012-2022)", fontsize=13, y=1.02)
    plt.tight_layout(); plt.show()
else:
    print("gini_coefficient not found")
```
