# AI-Driven Housing Affordability Forecasting in New York City
## IEEE-Standard Expanded Analysis: NTA-Level Panel Dataset (2012-2022)

**Dataset**: 2,512 rows x 49 columns covering 239 NTAs across 4 NYC boroughs  
**Target**: rent_burden_50plus_pct (HUD severely cost-burdened standard)  
**Methods**: Random Forest, XGBoost, LightGBM, Stacking Ensemble, SHAP, Ablation Study

## Section 3: Target Variable - Rent Burden Index

```python
# PRIMARY TARGET: rent_burden_50plus_pct
# SECONDARY: affordability_stress

df['affordability_stress'] = (
    df['median_gross_rent'] / df['renter_median_income'].clip(lower=1) * 12
)

# Winsorize extreme values in affordability_stress
q99 = df['affordability_stress'].quantile(0.99)
df['affordability_stress'] = df['affordability_stress'].clip(upper=q99)

print('=== Primary Target: rent_burden_50plus_pct ===')
print(df['rent_burden_50plus_pct'].describe().round(4))
print(f'Null count: {df["rent_burden_50plus_pct"].isnull().sum()}')
print()
print('=== Secondary Target: affordability_stress ===')
print(df['affordability_stress'].describe().round(4))
print(f'Null count: {df["affordability_stress"].isnull().sum()}')
```

## Section 4: Feature Engineering for IEEE

```python
# Market tightness: inverse of vacancy rate (clipped)
df['market_tightness'] = 1.0 / (df['vacancy_rate'].clip(lower=0.001) + 0.005)

# Rent-to-income ratio
df['rent_to_income_ratio'] = df['median_gross_rent'] * 12.0 / df['renter_median_income'].clip(lower=1)

# Housing burden composite
df['housing_burden_composite'] = df['unemployment_rate'] + df['severe_crowding_rate']

# Income inequality (already present as gini_coefficient)
df['income_inequality'] = df['gini_coefficient']

# Renter vulnerability index
df['renter_vulnerability'] = (
    (1.0 - df['renter_income_ratio'].clip(lower=0, upper=1)) * 0.5
    + df['unemployment_rate'] * 0.3
    + df['severe_crowding_rate'] * 0.2
)

# Verify new features
new_feats = ['market_tightness', 'rent_to_income_ratio', 'housing_burden_composite',
             'income_inequality', 'renter_vulnerability']
print('=== New Engineered Features ===')
print(df[new_feats].describe().round(4))
print()
print('=== Feature Null Counts ===')
print(df[new_feats].isnull().sum())
print()

# Correlation with target
target_col = 'rent_burden_50plus_pct'
corrs = df[new_feats + [target_col]].corr()[target_col].drop(target_col)
print('=== Correlation of New Features with Target ===')
print(corrs.sort_values(ascending=False).round(4))
```

```python
# Full correlation heatmap
corr_cols = [
    'median_hh_income', 'renter_median_income', 'median_gross_rent',
    'rent_burden_30plus_pct', 'rent_burden_50plus_pct', 'vacancy_rate',
    'unemployment_rate', 'severe_crowding_rate', 'gini_coefficient',
    'market_tightness', 'rent_to_income_ratio', 'renter_vulnerability'
]
available_corr = [c for c in corr_cols if c in df.columns]
corr_matrix = df[available_corr].corr().round(2)

fig, ax = plt.subplots(figsize=(13, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(
    corr_matrix, mask=mask, annot=True, fmt='.2f',
    cmap='RdBu_r', center=0, vmin=-1, vmax=1,
    linewidths=0.4, ax=ax, cbar_kws={'label': 'Pearson Correlation'}
)
ax.set_title('Fig 7: Correlation Matrix of Key Housing Market Features', fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'ieee_fig7_correlation_heatmap.png'), bbox_inches='tight')
plt.show()
print('Fig 7 saved.')
```

## Section 5: Train/Val/Test Split (IEEE Time-Series Cross-Validation)

```python
TARGET = 'rent_burden_50plus_pct'

FEATURES = [
    'median_hh_income', 'renter_median_income', 'renter_income_ratio', 'income_gap',
    'median_gross_rent', 'median_contract_rent', 'rent_burden_30plus_pct',
    'renter_share', 'homeownership_rate', 'vacancy_rate', 'unemployment_rate',
    'severe_crowding_rate', 'transit_commute_rate', 'gini_coefficient',
    'eviction_rate', 'market_tightness', 'rent_to_income_ratio',
    'housing_burden_composite', 'renter_vulnerability', 'income_growth_yoy',
    'rent_growth_yoy', 'median_hh_income_lag1', 'renter_median_income_lag1',
    'unemployment_rate_lag1', 'rent_burden_30plus_pct_lag1', 'vacancy_rate_lag1',
    'median_gross_rent_lag1', 'borough_code', 'year', 'covid_year'
]

# Only keep columns that exist in dataframe
FEATURES = [f for f in FEATURES if f in df.columns]
print(f'Features used: {len(FEATURES)}')
print(FEATURES)

# Drop rows where target is NaN
# Deduplicate column list (FEATURES already contains 'year')
_all_cols = list(dict.fromkeys(FEATURES + [TARGET, 'year', 'nta_name', 'borough_name']))
df_model = df[_all_cols].copy()
df_model = df_model.dropna(subset=[TARGET])
print(f'Rows after dropping NaN target: {len(df_model)}')

# Replace inf with NaN
df_model.replace([np.inf, -np.inf], np.nan, inplace=True)

# Train/Val/Test split by year
train_mask = df_model['year'] <= 2019
val_mask = df_model['year'] == 2020
test_mask = df_model['year'] >= 2021

train_df = df_model[train_mask].copy()
val_df = df_model[val_mask].copy()
test_df = df_model[test_mask].copy()

print(f'Train size: {len(train_df)} (years 2012-2019)')
print(f'Val size: {len(val_df)} (year 2020)')
print(f'Test size: {len(test_df)} (years 2021-2022)')

X_train = train_df[FEATURES].copy()
y_train = train_df[TARGET].values
X_val = val_df[FEATURES].copy()
y_val = val_df[TARGET].values
X_test = test_df[FEATURES].copy()
y_test = test_df[TARGET].values

# Impute
imputer = SimpleImputer(strategy='median')
X_train_imp = imputer.fit_transform(X_train)
X_val_imp = imputer.transform(X_val)
X_test_imp = imputer.transform(X_test)

# Combined train+val for final model training
X_trainval = np.vstack([X_train_imp, X_val_imp])
y_trainval = np.concatenate([y_train, y_val])

print(f'X_train shape: {X_train_imp.shape}')
print(f'X_test shape: {X_test_imp.shape}')
print(f'y_train range: [{y_train.min():.4f}, {y_train.max():.4f}]')
print(f'y_test range: [{y_test.min():.4f}, {y_test.max():.4f}]')
```

```python
# Time-series cross-validation setup
years_ordered = sorted(df_model['year'].unique().tolist())
print('All years for TSCV:', years_ordered)

tscv = TimeSeriesSplit(n_splits=5)
year_indices = df_model[FEATURES + [TARGET]].copy()
year_indices.replace([np.inf, -np.inf], np.nan, inplace=True)
year_indices_imputed = imputer.transform(year_indices[FEATURES])
year_target = year_indices[TARGET].values

print(f'TimeSeriesSplit with 5 folds configured.')
for fold_i, (tr_idx, te_idx) in enumerate(tscv.split(year_indices_imputed)):
    print(f'  Fold {fold_i+1}: Train size={len(tr_idx)}, Test size={len(te_idx)}')
```

## Section 6: Model Training

```python
def evaluate_model(model, X_tr, y_tr, X_te, y_te, model_name='Model'):
    """Evaluate model and print metrics."""
    pred_train = model.predict(X_tr)
    pred_test = model.predict(X_te)

    train_rmse = np.sqrt(mean_squared_error(y_tr, pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_te, pred_test))
    train_mae = mean_absolute_error(y_tr, pred_train)
    test_mae = mean_absolute_error(y_te, pred_test)
    train_r2 = r2_score(y_tr, pred_train)
    test_r2 = r2_score(y_te, pred_test)

    print(f'--- {model_name} ---')
    print(f'  Train RMSE: {train_rmse:.5f}  |  Test RMSE: {test_rmse:.5f}')
    print(f'  Train MAE:  {train_mae:.5f}  |  Test MAE:  {test_mae:.5f}')
    print(f'  Train R2:   {train_r2:.5f}  |  Test R2:   {test_r2:.5f}')

    return {
        'model': model_name,
        'train_rmse': train_rmse, 'test_rmse': test_rmse,
        'train_mae': train_mae, 'test_mae': test_mae,
        'train_r2': train_r2, 'test_r2': test_r2,
        'pred_test': pred_test
    }

model_results = []
print('Helper functions defined.')
```

```python
# Section 6a: Random Forest
print('Training Random Forest...')
rf_model = RandomForestRegressor(
    n_estimators=500,
    max_depth=12,
    min_samples_leaf=3,
    n_jobs=-1,
    random_state=42
)
rf_model.fit(X_trainval, y_trainval)
rf_results = evaluate_model(rf_model, X_trainval, y_trainval, X_test_imp, y_test, 'Random Forest')
model_results.append(rf_results)
print('Random Forest training complete.')
```

```python
# Section 6b: XGBoost
print('Training XGBoost...')
xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    n_jobs=-1,
    random_state=42,
    verbosity=0
)
xgb_model.fit(X_trainval, y_trainval)
xgb_results = evaluate_model(xgb_model, X_trainval, y_trainval, X_test_imp, y_test, 'XGBoost')
model_results.append(xgb_results)
print('XGBoost training complete.')
```

```python
# Section 6c: LightGBM
lgb_results = None
lgb_model = None

if LGB_AVAILABLE:
    print('Training LightGBM...')
    lgb_model = lgb.LGBMRegressor(
        n_estimators=500,
        max_depth=6,
        learning_rate=0.05,
        num_leaves=31,
        min_child_samples=20,
        subsample=0.8,
        n_jobs=-1,
        random_state=42,
        verbose=-1
    )
    lgb_model.fit(X_trainval, y_trainval)
    lgb_results = evaluate_model(lgb_model, X_trainval, y_trainval, X_test_imp, y_test, 'LightGBM')
    model_results.append(lgb_results)
    print('LightGBM training complete.')
else:
    print('LightGBM not available - skipping.')
```

```python
# Section 6d: Stacking Ensemble
print('Training Stacking Ensemble...')

base_estimators = [
    ('xgb', xgb.XGBRegressor(
        n_estimators=300, max_depth=6, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8, n_jobs=-1,
        random_state=42, verbosity=0
    )),
    ('rf', RandomForestRegressor(
        n_estimators=300, max_depth=12, min_samples_leaf=3,
        n_jobs=-1, random_state=42
    ))
]

if LGB_AVAILABLE and lgb_model is not None:
    base_estimators.append(('lgb', lgb.LGBMRegressor(
        n_estimators=300, max_depth=6, learning_rate=0.05,
        num_leaves=31, min_child_samples=20, subsample=0.8,
        n_jobs=-1, random_state=42, verbose=-1
    )))

stack_model = StackingRegressor(
    estimators=base_estimators,
    final_estimator=Ridge(alpha=1.0),
    cv=3,
    n_jobs=-1
)
stack_model.fit(X_trainval, y_trainval)
stack_results = evaluate_model(stack_model, X_trainval, y_trainval, X_test_imp, y_test, 'Stacking Ensemble')
model_results.append(stack_results)
print('Stacking Ensemble training complete.')
```

## Section 7: Time-Series Cross-Validation

```python
# Full dataset prepared for TSCV
df_full_cv = df_model[FEATURES + [TARGET]].copy()
df_full_cv.replace([np.inf, -np.inf], np.nan, inplace=True)

X_full = df_full_cv[FEATURES].values
y_full = df_full_cv[TARGET].values

# Re-impute on full dataset
imp_full = SimpleImputer(strategy='median')
X_full_imp = imp_full.fit_transform(X_full)

tscv5 = TimeSeriesSplit(n_splits=5)

cv_results = {'XGBoost': {'r2': [], 'rmse': []}, 'RandomForest': {'r2': [], 'rmse': []}}

for fold_i, (tr_idx, te_idx) in enumerate(tscv5.split(X_full_imp)):
    X_tr_cv, X_te_cv = X_full_imp[tr_idx], X_full_imp[te_idx]
    y_tr_cv, y_te_cv = y_full[tr_idx], y_full[te_idx]

    # XGBoost CV
    xgb_cv = xgb.XGBRegressor(
        n_estimators=300, max_depth=6, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        n_jobs=-1, random_state=42, verbosity=0
    )
    xgb_cv.fit(X_tr_cv, y_tr_cv)
    pred_cv = xgb_cv.predict(X_te_cv)
    cv_results['XGBoost']['r2'].append(r2_score(y_te_cv, pred_cv))
    cv_results['XGBoost']['rmse'].append(np.sqrt(mean_squared_error(y_te_cv, pred_cv)))

    # RF CV
    rf_cv = RandomForestRegressor(
        n_estimators=200, max_depth=12, min_samples_leaf=3,
        n_jobs=-1, random_state=42
    )
    rf_cv.fit(X_tr_cv, y_tr_cv)
    pred_rf_cv = rf_cv.predict(X_te_cv)
    cv_results['RandomForest']['r2'].append(r2_score(y_te_cv, pred_rf_cv))
    cv_results['RandomForest']['rmse'].append(np.sqrt(mean_squared_error(y_te_cv, pred_rf_cv)))

    print(f'Fold {fold_i+1}: XGB R2={cv_results["XGBoost"]["r2"][-1]:.4f}, RF R2={cv_results["RandomForest"]["r2"][-1]:.4f}')

print()
for model_nm in ['XGBoost', 'RandomForest']:
    r2_arr = cv_results[model_nm]['r2']
    rmse_arr = cv_results[model_nm]['rmse']
    print(f'{model_nm}: Mean R2={np.mean(r2_arr):.4f} +/- {np.std(r2_arr):.4f}, Mean RMSE={np.mean(rmse_arr):.5f} +/- {np.std(rmse_arr):.5f}')
```

## Section 9: Ablation Study (IEEE-Required)

```python
# Define feature groups for ablation
feature_groups = {
    'income_features': [
        'median_hh_income', 'renter_median_income', 'renter_income_ratio',
        'income_gap', 'income_growth_yoy', 'median_hh_income_lag1',
        'renter_median_income_lag1'
    ],
    'rental_market': [
        'median_gross_rent', 'median_contract_rent', 'rent_burden_30plus_pct',
        'rent_to_income_ratio', 'rent_growth_yoy', 'median_gross_rent_lag1',
        'rent_burden_30plus_pct_lag1'
    ],
    'labor_market': [
        'unemployment_rate', 'unemployment_rate_lag1',
        'housing_burden_composite'
    ],
    'housing_stock': [
        'renter_share', 'homeownership_rate', 'vacancy_rate',
        'severe_crowding_rate', 'market_tightness',
        'vacancy_rate_lag1', 'transit_commute_rate'
    ],
    'temporal_lags': [
        'median_hh_income_lag1', 'renter_median_income_lag1',
        'unemployment_rate_lag1', 'rent_burden_30plus_pct_lag1',
        'vacancy_rate_lag1', 'median_gross_rent_lag1'
    ],
    'spatial_identifiers': [
        'borough_code', 'year', 'covid_year'
    ]
}

# Filter to only features that exist in FEATURES list
for grp, feats in feature_groups.items():
    feature_groups[grp] = [f for f in feats if f in FEATURES]
    print(f'{grp}: {feature_groups[grp]}')

def train_eval_xgb(feats_subset, X_tr, y_tr, X_te, y_te):
    """Train XGBoost on a feature subset and return test R2."""
    feat_idx = [FEATURES.index(f) for f in feats_subset if f in FEATURES]
    if len(feat_idx) == 0:
        return 0.0
    X_tr_sub = X_tr[:, feat_idx]
    X_te_sub = X_te[:, feat_idx]
    m = xgb.XGBRegressor(
        n_estimators=300, max_depth=6, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        n_jobs=-1, random_state=42, verbosity=0
    )
    m.fit(X_tr_sub, y_tr)
    pred = m.predict(X_te_sub)
    return r2_score(y_te, pred)

print('Ablation study function defined.')
```

```python
# Run ablation study
print('Running ablation study...')

# Baseline: all features
baseline_r2 = xgb_results['test_r2']
print(f'Baseline R2 (all features): {baseline_r2:.5f}')

ablation_results = []

for grp_name, grp_feats in feature_groups.items():
    # Remove this group from features
    reduced_feats = [f for f in FEATURES if f not in grp_feats]

    if len(reduced_feats) == 0:
        ablation_r2 = 0.0
    else:
        ablation_r2 = train_eval_xgb(reduced_feats, X_trainval, y_trainval, X_test_imp, y_test)

    drop = baseline_r2 - ablation_r2
    ablation_results.append({
        'feature_group': grp_name,
        'num_features_removed': len(grp_feats),
        'test_r2_all': round(baseline_r2, 5),
        'test_r2_without': round(ablation_r2, 5),
        'r2_drop': round(drop, 5)
    })
    print(f'  Without {grp_name}: R2={ablation_r2:.5f} (drop={drop:.5f})')

ablation_df = pd.DataFrame(ablation_results).sort_values('r2_drop', ascending=False)
print('\n=== Ablation Study Results ===')
print(ablation_df.to_string(index=False))
```

```python
fig, ax = plt.subplots(figsize=(11, 6))

colors_abl = ['#C0392B' if d > 0 else '#2ECC71' for d in ablation_df['r2_drop']]
bars = ax.bar(ablation_df['feature_group'], ablation_df['r2_drop'],
              color=colors_abl, edgecolor='white', alpha=0.85)

ax.axhline(0, color='black', linewidth=1.2)
ax.set_xlabel('Feature Group Removed', fontsize=11)
ax.set_ylabel('R2 Drop (Baseline - Without Group)', fontsize=11)
ax.set_title('Fig 14: Ablation Study - R2 Drop When Removing Feature Groups (XGBoost)', fontsize=12)
plt.xticks(rotation=20, ha='right')

for bar, val in zip(bars, ablation_df['r2_drop']):
    h = bar.get_height()
    y_pos = h + 0.001 if h >= 0 else h - 0.003
    ax.text(bar.get_x() + bar.get_width() / 2, y_pos, f'{val:.4f}',
            ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'ieee_fig14_ablation_study.png'), bbox_inches='tight')
plt.show()
print('Fig 14 saved.')
```

## Section 10: Spatial Analysis - Moran's I

```python
# Compute model residuals
xgb_pred_test = xgb_results['pred_test']
residuals_test = y_test - xgb_pred_test

# Add residuals back to test dataframe
test_df_res = test_df.copy()
test_df_res['residual'] = residuals_test

# Borough-level mean residuals
boro_residuals = test_df_res.groupby('borough_name')['residual'].mean()
print('=== Borough-Level Mean Residuals ===')
print(boro_residuals.round(5))
print()

# Manual adjacency matrix for NYC boroughs
# Bronx - Manhattan (yes), Bronx - Queens (yes)
# Brooklyn - Manhattan (yes, via bridge), Brooklyn - Queens (yes)
# Manhattan - Bronx (yes), Manhattan - Brooklyn (yes), Manhattan - Queens (yes)
# Queens - Bronx (yes), Queens - Brooklyn (yes), Queens - Manhattan (yes)
borough_list_sorted = sorted(boro_residuals.index.tolist())
n_boro = len(borough_list_sorted)
print(f'Boroughs in test set: {borough_list_sorted}')

# Build adjacency (row-normalized W)
adjacency = {
    'Bronx':     ['Manhattan', 'Queens'],
    'Brooklyn':  ['Manhattan', 'Queens'],
    'Manhattan': ['Bronx', 'Brooklyn', 'Queens'],
    'Queens':    ['Bronx', 'Brooklyn', 'Manhattan']
}

W = np.zeros((n_boro, n_boro))
for i, b_i in enumerate(borough_list_sorted):
    neighbors = adjacency.get(b_i, [])
    for j, b_j in enumerate(borough_list_sorted):
        if b_j in neighbors:
            W[i, j] = 1.0

# Row-normalize
row_sums = W.sum(axis=1, keepdims=True)
row_sums[row_sums == 0] = 1
W_norm = W / row_sums

# Moran's I computation
res_vals = np.array([boro_residuals.get(b, 0.0) for b in borough_list_sorted])
res_mean = res_vals.mean()
z = res_vals - res_mean
n = len(z)

numerator = n * np.sum(W_norm * np.outer(z, z))
denominator = W_norm.sum() * np.sum(z ** 2)
morans_I = numerator / (denominator + 1e-12)

# Expected value and variance under null (random permutation)
E_I = -1.0 / (n - 1)
W_sum = W_norm.sum()
W_sym = W_norm + W_norm.T
S1 = 0.5 * np.sum(W_sym ** 2)
S2 = np.sum((W_norm.sum(axis=1) + W_norm.sum(axis=0)) ** 2)
S0 = W_sum

n_f = float(n)
b2 = (np.sum(z ** 4) / n_f) / ((np.sum(z ** 2) / n_f) ** 2)
Var_I = (
    n_f * ((n_f**2 - 3*n_f + 3) * S1 - n_f * S2 + 3 * S0**2)
    - b2 * ((n_f**2 - n_f) * S1 - 2*n_f * S2 + 6 * S0**2)
) / ((n_f - 1) * (n_f - 2) * (n_f - 3) * S0**2) - E_I**2

if Var_I > 0:
    z_score = (morans_I - E_I) / np.sqrt(Var_I)
else:
    z_score = 0.0

from scipy import stats
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print('=== Moran\'s I Spatial Autocorrelation Test ===')
print(f'Moran\'s I:       {morans_I:.5f}')
print(f'Expected I:      {E_I:.5f}')
print(f'Z-score:         {z_score:.4f}')
print(f'P-value:         {p_value:.4f}')
print()
if p_value < 0.05:
    if morans_I > E_I:
        print('RESULT: Significant POSITIVE spatial autocorrelation (p < 0.05).')
        print('Adjacent boroughs tend to have similar residual patterns.')
    else:
        print('RESULT: Significant NEGATIVE spatial autocorrelation (p < 0.05).')
        print('Adjacent boroughs tend to have dissimilar residual patterns.')
else:
    print('RESULT: No significant spatial autocorrelation in residuals (p >= 0.05).')
    print('Model residuals are spatially random at the borough level.')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Borough residual bar chart
colors_res = ['#C0392B' if r > 0 else '#2ECC71' for r in boro_residuals.values]
axes[0].bar(boro_residuals.index, boro_residuals.values, color=colors_res, edgecolor='white', alpha=0.8)
axes[0].axhline(0, color='black', linewidth=1.2)
axes[0].set_xlabel('Borough', fontsize=11)
axes[0].set_ylabel('Mean Residual (Actual - Predicted)', fontsize=11)
axes[0].set_title('Borough-Level Mean Model Residuals', fontsize=12)
axes[0].tick_params(axis='x', rotation=15)

# Residual distribution
axes[1].hist(residuals_test, bins=40, color='steelblue', edgecolor='white', alpha=0.8)
axes[1].axvline(0, color='red', linestyle='--', linewidth=2, label='Zero')
axes[1].axvline(residuals_test.mean(), color='orange', linestyle='--', linewidth=2,
                label=f'Mean: {residuals_test.mean():.4f}')
axes[1].set_xlabel('Residual', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].set_title('Distribution of Model Residuals (Test Set)', fontsize=12)
axes[1].legend()

plt.suptitle(
    "Fig 15: Spatial Analysis - Model Residuals\nMoran's I={:.4f}, p={:.4f}".format(morans_I, p_value),
    fontsize=12
)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'ieee_fig15_spatial_residuals.png'), bbox_inches='tight')
plt.show()
print('Fig 15 saved.')
```

```python
# NTAs that improved most vs worsened most (2012 to 2022)
nta_2012 = df[df['year'] == 2012][['nta_name', 'borough_name', 'rent_burden_50plus_pct']].set_index('nta_name')
nta_2022 = df[df['year'] == 2022][['nta_name', 'borough_name', 'rent_burden_50plus_pct']].set_index('nta_name')

# Only keep NTAs present in both years
common_ntas = nta_2012.index.intersection(nta_2022.index)
change_df = pd.DataFrame({
    'burden_2012': nta_2012.loc[common_ntas, 'rent_burden_50plus_pct'],
    'burden_2022': nta_2022.loc[common_ntas, 'rent_burden_50plus_pct'],
    'borough': nta_2012.loc[common_ntas, 'borough_name']
}).dropna()

change_df['change'] = change_df['burden_2022'] - change_df['burden_2012']

print('=== Top 10 Most Improved NTAs (2012 to 2022) ===')
improved = change_df.sort_values('change').head(10)
print(improved[['borough', 'burden_2012', 'burden_2022', 'change']].round(4).to_string())

print('\n=== Top 10 Most Worsened NTAs (2012 to 2022) ===')
worsened = change_df.sort_values('change', ascending=False).head(10)
print(worsened[['borough', 'burden_2012', 'burden_2022', 'change']].round(4).to_string())
```

```python
# NTA-level ranking for 2022
nta_ranking_2022 = year_2022.sort_values('rent_burden_50plus_pct', ascending=False).reset_index(drop=True)
nta_ranking_2022['rank'] = nta_ranking_2022.index + 1

print('=== Full NTA Ranking by Severe Rent Burden (2022) ===')
print(nta_ranking_2022[['rank', 'nta_name', 'borough_name', 'rent_burden_50plus_pct']].head(30).to_string(index=False))
```

```python
# Plot: Top 10 worst NTAs in 2022
top10_worst = nta_ranking_2022.head(10)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Worst NTAs 2022
colors_boro = {'Bronx': '#4472C4', 'Brooklyn': '#ED7D31', 'Manhattan': '#A9D18E', 'Queens': '#FF4444'}
bar_colors = [colors_boro.get(b, 'gray') for b in top10_worst['borough_name']]
axes[0].barh(top10_worst['nta_name'][::-1], top10_worst['rent_burden_50plus_pct'][::-1],
             color=bar_colors[::-1], edgecolor='white', alpha=0.85)
axes[0].set_xlabel('Rent Burden 50%+ (Share of Renters)', fontsize=10)
axes[0].set_title('Top 10 Worst NTAs by Severe Rent Burden (2022)', fontsize=11)
patches = [mpatches.Patch(color=v, label=k) for k, v in colors_boro.items()]
axes[0].legend(handles=patches, fontsize=8, loc='lower right')

# Most improved vs worsened
top5_improved = improved.head(5).copy()
top5_worsened = worsened.head(5).copy()
combined_change = pd.concat([
    top5_worsened.assign(label=lambda x: x.index + ' (+)'),
    top5_improved.assign(label=lambda x: x.index + ' (-)')
])
combined_change = combined_change.sort_values('change')
bar_cols2 = ['#2ECC71' if c < 0 else '#C0392B' for c in combined_change['change']]
axes[1].barh(combined_change.index, combined_change['change'],
             color=bar_cols2, edgecolor='white', alpha=0.85)
axes[1].axvline(0, color='black', linewidth=1.2)
axes[1].set_xlabel('Change in Rent Burden 50%+ (2012 to 2022)', fontsize=10)
axes[1].set_title('Most Improved vs Most Worsened NTAs (2012-2022)', fontsize=11)

plt.suptitle('Fig 16: Borough-Level Deep Dive - NTA Rent Burden Rankings', fontsize=13)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'ieee_fig16_borough_deepdive.png'), bbox_inches='tight')
plt.show()
print('Fig 16 saved.')
```

## Section 12: Forecasting (2023-2025)

```python
# Iterative forecast for 2023, 2024, 2025 at borough-median level
print('Generating 2023-2025 forecasts at borough-median level...')

# Compute borough-level medians for latest year (2022)
latest_year_data = df[df['year'] == 2022].copy()
borough_medians = latest_year_data.groupby('borough_name')[FEATURES + [TARGET]].median().reset_index()

# Historical borough medians for trend analysis
hist_trend = df.groupby(['borough_name', 'year'])[[TARGET]].median().reset_index()

forecast_rows = []
current_state = borough_medians.copy()

for forecast_year in [2023, 2024, 2025]:
    next_state = current_state.copy()
    next_state['year'] = forecast_year
    next_state['covid_year'] = 0  # Post-COVID

    # Update lag features from current state
    if 'median_hh_income_lag1' in FEATURES:
        next_state['median_hh_income_lag1'] = current_state.get('median_hh_income', current_state.get('median_hh_income_lag1', 0))
    if 'renter_median_income_lag1' in FEATURES:
        next_state['renter_median_income_lag1'] = current_state.get('renter_median_income', current_state.get('renter_median_income_lag1', 0))
    if 'unemployment_rate_lag1' in FEATURES:
        next_state['unemployment_rate_lag1'] = current_state.get('unemployment_rate', current_state.get('unemployment_rate_lag1', 0))
    if 'rent_burden_30plus_pct_lag1' in FEATURES:
        next_state['rent_burden_30plus_pct_lag1'] = current_state.get('rent_burden_30plus_pct', current_state.get('rent_burden_30plus_pct_lag1', 0))
    if 'vacancy_rate_lag1' in FEATURES:
        next_state['vacancy_rate_lag1'] = current_state.get('vacancy_rate', current_state.get('vacancy_rate_lag1', 0))
    if 'median_gross_rent_lag1' in FEATURES:
        next_state['median_gross_rent_lag1'] = current_state.get('median_gross_rent', current_state.get('median_gross_rent_lag1', 0))

    # Apply modest annual growth assumptions
    if 'median_hh_income' in next_state.columns:
        next_state['median_hh_income'] = next_state['median_hh_income'] * 1.025  # 2.5% income growth
    if 'renter_median_income' in next_state.columns:
        next_state['renter_median_income'] = next_state['renter_median_income'] * 1.02  # 2% renter income growth
    if 'median_gross_rent' in next_state.columns:
        next_state['median_gross_rent'] = next_state['median_gross_rent'] * 1.03  # 3% rent growth

    # Recompute engineered features
    if 'vacancy_rate' in next_state.columns:
        next_state['market_tightness'] = 1.0 / (next_state['vacancy_rate'].clip(lower=0.001) + 0.005)
    if 'renter_median_income' in next_state.columns and 'median_gross_rent' in next_state.columns:
        next_state['rent_to_income_ratio'] = next_state['median_gross_rent'] * 12.0 / next_state['renter_median_income'].clip(lower=1)
    if 'unemployment_rate' in next_state.columns and 'severe_crowding_rate' in next_state.columns:
        next_state['housing_burden_composite'] = next_state['unemployment_rate'] + next_state['severe_crowding_rate']
    if 'renter_income_ratio' in next_state.columns:
        next_state['renter_vulnerability'] = (
            (1.0 - next_state['renter_income_ratio'].clip(lower=0, upper=1)) * 0.5
            + next_state.get('unemployment_rate', 0) * 0.3
            + next_state.get('severe_crowding_rate', 0) * 0.2
        )

    # Predict
    X_forecast = next_state[FEATURES].copy()
    X_forecast.replace([np.inf, -np.inf], np.nan, inplace=True)
    X_forecast_imp = imputer.transform(X_forecast)
    preds = xgb_model.predict(X_forecast_imp)

    for idx, (_, row) in enumerate(next_state.iterrows()):
        forecast_rows.append({
            'borough_name': row['borough_name'],
            'year': forecast_year,
            'predicted_rent_burden_50plus': round(float(preds[idx]), 5)
        })

    current_state = next_state.copy()
    print(f'  {forecast_year}: predictions generated.')

forecast_df = pd.DataFrame(forecast_rows)
print('\n=== Forecast Results (Borough-Level) ===')
print(forecast_df.pivot(index='borough_name', columns='year', values='predicted_rent_burden_50plus').round(4))
```

```python
fig, ax = plt.subplots(figsize=(13, 7))

boroughs_fc = sorted(forecast_df['borough_name'].unique().tolist())

for i, boro in enumerate(boroughs_fc):
    # Historical
    hist_boro = hist_trend[hist_trend['borough_name'] == boro].sort_values('year')
    ax.plot(hist_boro['year'], hist_boro[TARGET],
            marker='o', linewidth=2, markersize=5,
            color=colors_list[i % len(colors_list)], label=boro)

    # Forecast
    fc_boro = forecast_df[forecast_df['borough_name'] == boro].sort_values('year')
    # Connect last historical point to first forecast point
    last_hist_year = hist_boro['year'].max()
    last_hist_val = hist_boro[hist_boro['year'] == last_hist_year][TARGET].values[0]
    fc_years = [last_hist_year] + fc_boro['year'].tolist()
    fc_vals = [last_hist_val] + fc_boro['predicted_rent_burden_50plus'].tolist()

    ax.plot(fc_years, fc_vals,
            marker='D', linewidth=2, markersize=6, linestyle='--',
            color=colors_list[i % len(colors_list)], alpha=0.8)

# Shading
ax.axvspan(2019.5, 2021.5, alpha=0.12, color='red', label='COVID-19 Period')
ax.axvspan(2022.5, 2025.5, alpha=0.08, color='blue', label='Forecast Zone (2023-2025)')
ax.axvline(x=2022.5, color='blue', linestyle=':', linewidth=1.8)

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Rent Burden 50%+ (Share of Renters)', fontsize=12)
ax.set_title('Fig 17: Borough-Level Rent Burden Forecast (Historical 2012-2022 + Forecast 2023-2025)', fontsize=12)
ax.legend(loc='upper right', fontsize=9)
all_years = list(range(2012, 2026))
ax.set_xticks(all_years)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'ieee_fig17_forecast.png'), bbox_inches='tight')
plt.show()
print('Fig 17 saved.')
```
